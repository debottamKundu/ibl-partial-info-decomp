```python
def find_minimum_cover_with_dropping(animal_region_map, target_set):
    selected_animals = set()
    uncovered_regions = set(target_set)
    
    while uncovered_regions:
        # Select the animal that covers the most uncovered regions
        best_animal = None
        best_cover = set()
        
        for animal, regions in animal_region_map.items():
            covered = regions & uncovered_regions  # intersect with uncovered regions
            if len(covered) > len(best_cover):
                best_animal = animal
                best_cover = covered
        
        if not best_cover:
            # No animals can cover any of the remaining uncovered regions
            print(f"Warning: Could not cover the remaining regions: {uncovered_regions}")
            break
        
        # Add this animal to the selected set
        selected_animals.add(best_animal)
        # Remove the covered regions from the uncovered set
        uncovered_regions -= best_cover

    # Check if all regions in L are covered
    covered_regions = set.union(*[animal_region_map[a] for a in selected_animals])
    uncovered_regions_final = target_set - covered_regions

    if not uncovered_regions_final:
        return selected_animals
    else:
        print(f"Could not cover regions: {uncovered_regions_final}. Attempting to drop them.")
        # Try dropping the uncovered regions and find the minimum cover for the remaining regions
        remaining_regions = target_set - uncovered_regions_final
        return selected_animals, uncovered_regions_final
```

```python
# just drop lateral in the analysis
minimum_cover, regions_not_covered = find_minimum_cover_with_dropping(animal_region_map, set(areas_to_filter))
```

```python
for eid in complete_set:
    pids, probes = one.eid2pid(eid)
    if isinstance(probes, list) and len(probes) > 1:
        x = [load_good_units(one, pid=None, eid=eid, qc=1, pname=probe_name)
                    for probe_name in probes]
    else:   
        spikes, clusters = load_good_units(one, pid=None, eid=eid, qc=1, pname=probes)
```

```python
min_units = 10
for region in regions:
    # find all clusters in region (where region can be a list of regions)
    region_mask = np.isin(beryl_regions, region)
    if sum(region_mask) < 10:
        # 
        continue
    else:
        print(f'Region found {region}, {np.sum(region_mask)}')
    
#then get data from region mask
binned_spikes = []
actual_regions = []
n_units = []
cluster_uuids_list = []
for region in regions:
    # find all clusters in region (where region can be a list of regions)
    region_mask = np.isin(beryl_regions, region)
    if sum(region_mask) < min_units:
        print(f"{(region)} below min units threshold ({min_units}) : {sum(region_mask)}, not decoding")
    else:
        # find all spikes in those clusters
        spike_mask = np.isin(spikes['clusters'], clusters[region_mask].index)
        times_masked = spikes['times'][spike_mask]
        clusters_masked = spikes['clusters'][spike_mask]
        # record cluster uuids
        idxs_used = np.unique(clusters_masked)
        clusters_uuids = list(clusters.iloc[idxs_used]['uuids'])
        # bin spikes from those clusters
        
        binned, _ = get_spike_counts_in_bins(spike_times=times_masked, spike_clusters=clusters_masked, intervals=intervals)
        binned = binned.T
        binned_spikes.append(binned)
        actual_regions.append(region)
        n_units.append(sum(region_mask))
        cluster_uuids_list.append(clusters_uuids)
```
