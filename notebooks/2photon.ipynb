1. To compare like-for-like, you will want to match the fields-of-view (FOVs) you are looking at, across days.
2. To do this, for a given session in a given mouse, query what the location is of each FOV's centroid by taking the geometric mean of its pixel-wise mlapdv location estimate, found in mpciMeanImage.mlapdv_estimate.npy.
3. Then, for all other sessions in that mouse, ask whether there are any FOV's that have matching centroid locations.
4. With this approach you will not be explicitly matching each neuron across days, but at least you will be comparing largely overlapping populations

```python
import pandas as pd
import pickle as pkl
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
```

```python
from glob import glob
```

```python
files = glob("../data/processed/twop/*.pkl")
```

```python
with open("../data/processed/twop/SP060.pkl", "rb") as f:
    data = pkl.load(f)
```

```python
def training_protocol(data, name):

    sessions_data = data["contrasts"]
    protocols = data["protocol"]

    all_contrasts = np.concatenate(sessions_data)
    unique_contrasts = np.unique(all_contrasts)
    num_unique = len(unique_contrasts)
    num_sessions = len(sessions_data)

    heatmap_matrix = np.zeros((num_unique, num_sessions))

    # Populate the matrix
    for session_idx, (contrasts, s_type) in enumerate(zip(sessions_data, protocols)):
        # Find the row indices for this session's contrasts
        row_indices = np.searchsorted(unique_contrasts, contrasts)
        if "training" in s_type.lower():
            heatmap_matrix[row_indices, session_idx] = 1
        elif "biased" in s_type.lower():
            heatmap_matrix[row_indices, session_idx] = 3
        elif "advanced" in s_type.lower():
            heatmap_matrix[row_indices, session_idx] = 2
        else:
            print(s_type)

    # 4. Plot using Seaborn
    plt.figure(figsize=(4, 4))

    # Pass the raw numpy array to seaborn
    # xticklabels and yticklabels map our specific names/values to the axes
    ax = sns.heatmap(
        heatmap_matrix,
        cmap="Blues",
        cbar=False,  # Hides the colorbar since it's just 0s and 1s
        linewidths=0.5,  # Adds a grid for better readability
        linecolor="lightgray",
        yticklabels=unique_contrasts,
        vmin=0,
        vmax=3,
    )

    # By default, heatmaps plot the first row (index 0) at the top.
    # Since our contrasts are sorted smallest to largest, we invert the Y-axis
    # so the lowest floats start at the bottom, like a standard graph.
    ax.invert_yaxis()

    plt.title(f"Subject {name}")
    plt.xlabel("Session")
    plt.ylabel("Contrasts")

    plt.tight_layout()
    plt.show()
```

```python
for fname in files:
    name = fname.rsplit("/")[-1].rstrip(".pkl")
    with open(fname, "rb") as f:
        data = pkl.load(f)
    training_protocol(data, name)
```
