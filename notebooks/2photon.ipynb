1. To compare like-for-like, you will want to match the fields-of-view (FOVs) you are looking at, across days.
2. To do this, for a given session in a given mouse, query what the location is of each FOV's centroid by taking the geometric mean of its pixel-wise mlapdv location estimate, found in mpciMeanImage.mlapdv_estimate.npy.
3. Then, for all other sessions in that mouse, ask whether there are any FOV's that have matching centroid locations.
4. With this approach you will not be explicitly matching each neuron across days, but at least you will be comparing largely overlapping populations



```python
from one.api import ONE
import numpy as np

one = ONE()
```

```python
trained_animals = one.search(task_protocol="biasedChoiceWorld", procedures="Imaging")
in_training_animals = one.search(task_protocol="trainingChoiceWorld", procedures="Imaging")
```

```python
training_subjects = []
trained_subjects = []

for idx in range(len(in_training_animals)):
    training_subjects.append(one.get_details(in_training_animals[idx])["subject"])


for idx in range(len(trained_animals)):
    training_subjects.append(one.get_details(trained_animals[idx])["subject"])
```

```python
training_subjects = np.unique(training_subjects)
trained_subjects = np.unique(trained_subjects)
print(training_subjects)
print(trained_subjects)
```
